```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained("mistralai/Mistral-7B-Instruct-v0.2")

# Load quantized model
model = AutoModelForCausalLM.from_pretrained(
    "mistralai/Mistral-7B-Instruct-v0.2",
    device_map="auto",  # Automatically map layers to GPUs
    load_in_8bit=True,  # Enable 8-bit quantization
    torch_dtype=torch.float16  # Use FP16 precision for speed and memory efficiency
)

# Initialize pipeline
llm_pipeline = pipeline("text-generation", model=model, tokenizer=tokenizer)

# # Example usage
# output = llm_pipeline("What is quantization?", max_length=50)
# print(output)
```

```python
# Prompt for table summarization
prompt = (
    "Summarize the following table data:\n\n"
    f"{table_text}\n\n"
    "Provide a concise summary of the key points."
)

# Generate summary using max_new_tokens instead of max_length
response = llm_pipeline(prompt, max_new_tokens=100, num_return_sequences=1)
summary = response[0]['generated_text']
print(f"Summary:\n{summary}")
```
